```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.animation import FuncAnimation
import time
from IPython.display import HTML
```

```python
def insert_sort_with_steps(arr):
    steps = [arr.copy()]  # 记录初始状态
    operations = ["初始数组状态"]
    keys = [-1]  # 初始无key
    for i in range(1, len(arr)):
        key = arr[i]
        operations.append(f"选择 key={key} (位置{i}) 开始插入")
        steps.append(arr.copy())
        keys.append(i)  # key位置
        j = i - 1
        while j >= 0 and key < arr[j]:
            operations.append(f"key={key} < arr[{j}]={arr[j]}，移动 arr[{j}] 到 arr[{j+1}]")
            arr[j + 1] = arr[j]
            steps.append(arr.copy())  # 记录移动后的状态
            keys.append(i)  # key仍在i位置
            j -= 1
        operations.append(f"插入 key={key} 到位置 {j+1}")
        arr[j + 1] = key
        steps.append(arr.copy())  # 记录插入后的状态
        keys.append(-1)  # 插入完成，无key
    return steps, operations, keys
```

```python
def update(frame):
    current_arr = steps[frame]
    for i, (bar, height) in enumerate(zip(bars, current_arr)):
        bar.set_height(height)
        texts[i].set_text(str(height))
        texts[i].set_position((i, height + 0.5))
    ax.set_title(f'步骤 {frame + 1}/{len(steps)}: {operations[frame]}')
    return list(bars) + texts
```

```python
# 测试数组
test_array = [12, 11, 13, 5, 6]
steps, operations, keys = insert_sort_with_steps(test_array.copy())  # 使用副本避免修改原数组

# 设置动画
fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.bar(range(len(steps[0])), steps[0], color='skyblue')
ax.set_ylim(0, max(test_array) + 5)
ax.set_xticks(range(len(test_array)))
ax.set_xticklabels([f'位置{i}' for i in range(len(test_array))])
ax.set_ylabel('数值')
ax.set_xlabel('数组位置')

# 添加数值标签
texts = [ax.text(i, steps[0][i] + 0.5, str(steps[0][i]), ha='center', va='bottom', fontsize=12) for i in range(len(steps[0]))]

def update(frame):
    current_arr = steps[frame]
    key_idx = keys[frame]
    for i, (bar, height) in enumerate(zip(bars, current_arr)):
        bar.set_height(height)
        if i == key_idx:
            bar.set_color('red')
        else:
            bar.set_color('skyblue')
        texts[i].set_text(str(height))
        texts[i].set_position((i, height + 0.5))
    ax.set_title(f'步骤 {frame + 1}/{len(steps)}: {operations[frame]}')
    return list(bars) + texts

ani = FuncAnimation(fig, update, frames=len(steps), interval=2000, repeat=False)
plt.close(fig)  # 关闭静态图
display(HTML(ani.to_jshtml()))  # 显示动画
```
